### 3.7 Prob 24

Define $R(x) = -\log{|\overline{F}(x)|}$. Then with $\overline{F}(x) = 1 - F(x)$, we may drop the absolute values (as $F(x) \leq 1$) and differentiate to obtain:

\begin{align}
    \frac{dR}{dx} & = \frac{-1}{1-F(x)}\cdot\frac{d}{dx}[1-F(x)] &(\text{chain rule})\\
        & = \frac{-1}{1-F(x)}\cdot[-f(x)] & (\text{Fundemental Theorem of Calculus})\\
        & = \frac{f(x)}{1-F(x)}\\
        & = r(x) &(\text{definition of the hazard function})
  \end{align}

### 3.7 Prob 30

**Lemma.** If $X \sim \Gamma(1,\mu)$, with $\mu > 0$, and $Y = \frac X\mu$, then $Y \sim \mathrm{Exp}(1)$.

**Proof.**
\begin{align}
    \PP{Y \leq y} & = \PP{\frac X\mu \leq y}\\
        & = \PP{ X \leq \mu y}\\
        &= \int_0^{\mu y} f_X(t)\,dt\\
        &= \int_0^{\mu y} \frac1\mu e^{-t/\mu}\,dt\\
        &= \int_0^{-y} -e^u\, du &(\text{changing variables to } u = -t/\mu)\\
        & = 1 - e^{-y}.
  \end{align}.

**Theorem** ($n$th moment of $\Gamma(1, \mu)$ distribution). Let $X \sim \Gamma(1,\mu)$, with $\mu > 0$. I claim that 

$$ E[X^n] = n!\mu^n, \quad \text{for all natural numbers } n.$$

**Proof.** Suppose $X \sim \Gamma(1,\mu)$, $\mu > 0$, and $Y = \frac X\mu$. We know $Y \sim \mathrm{Exp}(1)$. We'll evaluate $E[X^n] = E[(\mu Y)^n]$. 

(*Base case*, $n=1$) 
\begin{align}
    E[\mu Y] &= \mu E[Y] & (\text{linearity})\\
            & = \mu \int_0^\infty t e^{-t} \,dt\\
        & = \mu\left[-t e^{-t}\bigg|_0^\infty + \int_0^\infty e^{-t} \,dt\right] & (\text{integration by parts})\\
        & = \mu \left[-e^{-t}\right]_0^\infty\\
        & = \mu.
  \end{align}

(*Inductive step*) Suppose $E[(\mu Y)^{n-1}] = (n-1)!\mu^{n-1}$. 

Then \begin{align}
    E[(\mu Y)^n] &= \mu^n \int_0^\infty t^n e{-t} \,dt\\
        & = \mu^n\left[-t^n e^{-t}\bigg|_0^\infty + \int_0^\infty n t^{n-1} e^{-t} \,dt\right] & (\text{integration by parts})\\
        & = \mu^n\left[\frac n{\mu^{n-1}} \mu^{n-1}\int_0^\infty t^{n-1} e^{-t}\,dt\right]\\
        & = \mu^n\left[\frac n{\mu^{n-1}} E[(\mu Y)^{n-1}]\right]\\
        & = \mu^n\frac n{\mu^{n-1}} \mu^{n-1}(n-1)! & (\text{inductive hypothesis})\\
        & = \mu^n n!.
        \end{align}

### 3.7 Prob 48

I claim if $X \sim N(\mu, \sigma^2)$ and $Y = \alpha X + \beta$ with $\alpha > 0$, then $Y \sim N(\alpha\mu + \beta, (\sigma\alpha)^2)$.

**Proof.** Suppose $X$ and $Y$ are as above. We'll show the cdf of $Y$ describes $N(\alpha\mu + \beta, (\sigma\alpha)^2)$.

\begin{align}
    F_Y(t) & = \PP{Y \leq y}\\
      & = \PP{X \leq \frac{t-\beta}{\alpha}}
          & (\text{set equality})\\
      & = \int_{-\infty}^{(t-\beta)/\alpha} \frac1{\sqrt{2\pi}\sigma}\exp{\frac{-(t'-\mu)^2}{2\sigma^2}}\, dt' 
          & (\text{cdf of $X$})\\
      & = \int_{-\infty}^t \frac1{\sqrt{2\pi}\sigma}\exp{\frac{-(\frac{u-\beta}{\alpha}-\mu)^2}{2\sigma^2}}\, \frac{du}\alpha 
          & (\text{change of variables $u = \alpha t' + \beta$})\\
      & = \int_{-\infty}^t \frac1{\sqrt{2\pi}\sigma\alpha}\exp{\frac{-(u-(\alpha\mu +\beta))^2}{2(\sigma\alpha)^2}}\, du
          & (\text{refactoring})\\
      & = F(t) & \text{which is the cdf of $N(\alpha\mu + \beta, (\sigma\alpha)^2)$}.
      \end{align}

We have identical cdfs iff we have indentical distributions.

**Corrolary.** If $X \sim N(\mu, \sigma^2)$, then $\frac 1\sigma X - \frac\mu\sigma \sim N(0,1)$, which is seen with $\alpha = \frac1\sigma$ and $\beta = -\frac\mu\sigma$.
